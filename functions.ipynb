```python
from time import *

def lunch_time(time):
    if time == 13:  # Assuming lunch is at 12 PM
        print("Will go for lunch")
    else:
        print("I will wait then we can go for lunch")

# Ask user for input and call the function
time_input = input("Enter the time (in 24-hour format): ")

# Check if input is valid (integer and between 0 and 23)
try:
    time_input = int(time_input)
    if 0 <= time_input <= 23:
        lunch_time(time_input)
    else:
        print("Please enter a valid time between 0 and 23.")
except ValueError:
    print("Please enter a valid integer for the time.")
```
